```python
import xarray as xr
import pandas as pd
import netCDF4
import netCDF4
import h5netcdf
```

```python
# read 003
ds_003=xr.open_dataset("D:/Man/Earth and Environment/project2/003_2006_2080_352_360.nc")
ds_003.to_dataframe()
```

```python
# select all the data from Manchester and change it to csv
man_lat=53.48
man_lon=357.76
ds_003_man=ds_003.sel(lat=man_lat,lon=man_lon,method='nearest')

if hasattr(ds_003_man.indexes['time'], 'to_datetimeindex'):
    datetime_index = ds_003_man.indexes['time'].to_datetimeindex()
    ds_003_man['time'] = datetime_index

man_003=ds_003_man.to_dataframe().reset_index()

man_003['year']=man_003['time'].dt.year
man_003['month']=man_003['time'].dt.month
man_003['day']=man_003['time'].dt.day_of_year

man_003.to_csv("D:/Man/Earth and Environment/project2/003_manchester.csv",index=False)
```
