```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read predicted dataset
df_003=pd.read_csv("D:/Man/Earth and Environment/prediction/dataset_03_predictions.csv")
df_003['time']=pd.to_datetime(df_003['time'])

df_004=pd.read_csv("D:/Man/Earth and Environment/prediction/dataset_04_predictions.csv")
df_004['time']=pd.to_datetime(df_004['time'])

df_006=pd.read_csv("D:/Man/Earth and Environment/prediction/dataset_06_predictions.csv")
df_006['time']=pd.to_datetime(df_006['time'])

df_008=pd.read_csv("D:/Man/Earth and Environment/prediction/dataset_08_predictions.csv")
df_008['time']=pd.to_datetime(df_008['time'])

```

```python
# TREFMXAV_U Time-series plot 003
yearly_mean_003=df_003.groupby("year")['TREFMXAV_U'].mean()

plt.plot(df_003['time'],df_003['TREFMXAV_U'],color="#8FD0F8")
plt.plot(yearly_mean_003.index.map(lambda x: pd.Timestamp(year=x, month=7, day=2)),
         yearly_mean_003.values, color="#F8CE8F", linewidth=2.5, label='Yearly Average Trend')
plt.title('Manchester Temperature Evolution (2050-2080/003)', fontsize=14)
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Merge datasets
data=[df_003,df_004,df_006,df_008]
labels=['003','004','006','008']
processed_dataset=[]
for df,label in zip(data,labels):
    temp_df=df.copy()
    temp_df['member_Id']=label
    processed_dataset.append(temp_df)

combined_df=pd.concat(processed_dataset,ignore_index=True)

combined_df.head()
```

```python
# TREFMXAV_U KDE
plt.figure(figsize=(10,6))

sns.kdeplot(data=combined_df,x='TREFMXAV_U',hue='member_Id',common_norm=False,palette='RdBu',linewidth=2)

plt.title('TREFMXAV_U Distribution (2050-2080)',fontsize=14,fontweight='bold')
plt.xlabel('TREFMXAV_U')
plt.ylabel('Probability Density')

plt.show()
```

```python
# TREFMXAV_U Boxplot
plt.figure(figsize=(12,7))
sns.boxplot(data=combined_df,x='member_Id',y='TREFMXAV_U',hue='member_Id',palette='RdBu',fliersize=3)

plt.title('TREFMXAV_U Distribution Comparison (2050-2080)',fontsize=14,fontweight='bold')
plt.xlabel('Member',fontsize=12)
plt.ylabel('TREFMXAV_U',fontsize=12)
plt.grid(alpha=0.5)

plt.show()
```

```python
# TREFMXAV_U Boxplot
plt.figure(figsize=(12,7))
sns.boxplot(data=df_003,x='member_Id',y='TREFMXAV_U',palette='RdBu',fliersize=3)

plt.title('TREFMXAV_U Distribution Comparison (003)',fontsize=14,fontweight='bold')
plt.xlabel('Member',fontsize=12)
plt.ylabel('TREFMXAV_U',fontsize=12)
plt.grid(alpha=0.5)

plt.show()
```

```python
# Yearly Temperature Trend
yearly_avg_combined = combined_df.groupby(['member_Id', 'year'])['TREFMXAV_U'].mean().reset_index()

plt.figure(figsize=(14, 7))

sns.lineplot(data=yearly_avg_combined,x='year',y='TREFMXAV_U',hue='member_Id',marker='o',markersize=5,linewidth=2,palette='RdBu')

plt.title('Annual Average Temperature Trend Comparison', fontsize=15, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Member ID', title_fontsize='12', loc='upper left')

plt.show()
```
