```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read raw dataset 003-008
read_003=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/003_manchester.csv")
read_004=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/004_manchester.csv")
read_005=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/005_manchester.csv")
read_006=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/006_manchester.csv")
read_007=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/007_manchester.csv")
read_008=pd.read_csv("D:/Man/Earth and Environment/EART60702_project2_group3/dataset/008_manchester.csv")
```

```python
# TREFMXAV_U Time-series plot 003
read_003['time']=pd.to_datetime(read_003['time'])
temp_003=read_003.copy()
temp_003['TREFMXAV_U']=temp_003['TREFMXAV_U']-273.15
yearly_mean=temp_003.groupby("year")['TREFMXAV_U'].mean()

plt.plot(temp_003['time'],temp_003['TREFMXAV_U'])
plt.plot(yearly_mean.index.map(lambda x: pd.Timestamp(year=x, month=7, day=2)),
         yearly_mean.values, color='#FF9800', linewidth=2.5, label='Yearly Average Trend')
plt.title('Manchester Temperature Evolution (2006-2080)', fontsize=16)
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# TREFMXAV_U Time-series plot 004
read_004['time']=pd.to_datetime(read_004['time'])
temp_004=read_004.copy()
temp_004['TREFMXAV_U']=temp_004['TREFMXAV_U']-273.15
yearly_mean_004=temp_004.groupby("year")['TREFMXAV_U'].mean()

plt.plot(temp_004['time'],temp_004['TREFMXAV_U'])
plt.plot(yearly_mean_004.index.map(lambda x: pd.Timestamp(year=x, month=7, day=2)),
         yearly_mean_004.values, color='#FF9800', linewidth=2.5, label='Yearly Average Trend')
plt.title('Manchester Temperature Evolution (2006-2080)', fontsize=16)
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Merge datasets
data=[read_003,read_004,read_005,read_006,read_007,read_008]
labels=['003','004','005','006','007','008']
processed_dataset=[]
for df,label in zip(data,labels):
    temp_df=df.copy()
    temp_df['TREFMXAV_U']=temp_df['TREFMXAV_U']-273.15
    temp_df['TREFHT']=temp_df['TREFHT']-273.15
    temp_df['member_Id']=label
    processed_dataset.append(temp_df)

combined_df=pd.concat(processed_dataset,ignore_index=True)

combined_df.head()
```

```python
# TREFMXAV_U KDE
plt.figure(figsize=(10,6))

sns.kdeplot(data=combined_df,x='TREFMXAV_U',hue='member_Id',common_norm=False,palette='RdBu',linewidth=2)

plt.title('TREFMXAV_U Distribution (003-008)',fontsize=14,fontweight='bold')
plt.xlabel('TREFMXAV_U')
plt.ylabel('Probability Density')

plt.show()
```

```python
# TREFMXAV_U Boxplot
plt.figure(figsize=(12,7))
sns.boxplot(data=combined_df,x='member_Id',y='TREFMXAV_U',hue='member_Id',palette='RdBu',fliersize=3)

plt.title('TREFMXAV_U Distribution Comparison (003-008)',fontsize=14,fontweight='bold')
plt.xlabel('Member',fontsize=12)
plt.ylabel('TREFMXAV_U',fontsize=12)
plt.grid(alpha=0.5)

plt.show()
```
