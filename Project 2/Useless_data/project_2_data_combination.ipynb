```python
import xarray as xr
import pandas as pd
import os
```

```python
# Step 1: Read File
folder_path = r"Data"

file_list = [
    "003_2006_2080_352_360.nc",
    "004_2006_2080_352_360.nc",
    "005_2006_2080_352_360.nc",
    "006_2006_2080_352_360.nc",
    "007_2006_2080_352_360.nc",
    "008_2006_2080_352_360.nc"
]
```

```python
# Step 2: Add & Count memberid
df_list = []

for file in file_list:
    file_path = os.path.join(folder_path, file)

    ds = xr.open_dataset(file_path)
    df_temp = ds.to_dataframe().reset_index()

    member_id = int(file[:3])
    df_temp['member_id'] = member_id

    df_list.append(df_temp)

df = pd.concat(df_list, ignore_index=True)

print("Final line number:", len(df))
print("six member_id:")
print(df['member_id'].value_counts())
```

```python
# Step 3: Combination

df = pd.concat(df_list, ignore_index=True)

print("Combination Finish")
print(df.head())
```

```python
# Step 4: Generate Output csv
output_path = r"C:\Users\Danya\Desktop\Project 2\final_merged_data.csv"

df.to_csv(output_path, index=False)

print("CSV Created")
```
